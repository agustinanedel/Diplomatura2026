##Resolución Ejercicio 1


1. Seleccionamos las columnas relevantes a analizar


*   **tools_programming_languages:** porque es la variable predictora. Es la columna que contiene la información central de la pregunta.

*  **Salary_monthly_neto:** es la variable de respuesta. Se prefiere el salario neto sobre el bruto porque refleja la realidad del bolsillo. Mientras que el bruto incluye cargas sociales e impuestos que varían.


*   **work_seniority:** considera la experiencia, permite validar si un lenguaje es bien pagado incluso para alguien que recien empieza. En este caso, se emplea como validación datos porque puedo saber si realmente es un lenguaje bien remunerado en todos los niveles.  







```
# 1. Definimos la  lista con los nombres  de las 3 columnas
columnas_elegidas = ['tools_programming_languages', 'salary_monthly_NETO', 'work_seniority']

# 2. Creamos una nueva tabla que solo tenga esas columnas
# Usamos .copy() para que sea una tabla independiente y no afecte a la original
df_seleccion = df[columnas_elegidas].copy()

# 3. Verificamos el resultado mostrando las primeras 10 filas
print(f"Registros después de la selección de columnas: {len(df_seleccion)}")
df_seleccion.head(10)
```

2. Selección de las filas más relevantes:


a) **Filtro de Salario**: se limitó el análisis a personas con salarios netos entre 300.000 y 10.000.000. El límite inferior descarta pasantías, errores de carga que no representan el mercado laboral estándar de IT. El superior es para evitar outliers o valores atípicos, que distorsionen el promedio.




```
# 1. Definimos los límites
limite_inferior = 300000
limite_superior = 10000000

# 2. El símbolo '&' significa que se deben cumplir AMBAS condiciones
df_filtrado = df_seleccion[(df_seleccion['salary_monthly_NETO'] >= limite_inferior) &
                           (df_seleccion['salary_monthly_NETO'] <= limite_superior)]

# 3. Se eliminan las filas que no tengan ningún dato en el salario (NaN)
df_filtrado = df_filtrado.dropna(subset=['salary_monthly_NETO'])

# 4. Resultados
print(f"Registros después de filtrar valores extremos: {len(df_filtrado)}")
df_filtrado.head(10)
```

b)   **Selección de Lenguajes:** se seleccionó 10 lenguajes más frecuentes, como Python, SQL, JavaScript, etc, para asegurar que cada grupo tenga suficientes datos y ser estadísticamente representativo, ya que se necesita "N" (Cantidad de personas) mínima para que el dato sea válido. Si analizados un lenguaje que solo usa 1 persona, ese sueldo representa a esa persona y no al lenguaje en si.

```
# 1. Para convertir los lenguajes separados por coma en una lista
def split_languages(languages_str):
  if not isinstance(languages_str, str):
    return []
  languages_str = languages_str.lower().replace('ninguno de los anteriores', '')
  return [lang.strip().replace(',', '') for lang in languages_str.split()]

# 2. Se aplica la función a la  tabla filtrada
df_filtrado.loc[:, 'cured_languages'] = df_filtrado['tools_programming_languages'].apply(split_languages)

# 3. Se separan los lenguajes en filas individuales
df_langs = df_filtrado.explode('cured_languages')

# 4. Se limpian filas vacías
df_langs = df_langs.dropna(subset=['cured_languages'])
df_langs = df_langs[df_langs['cured_languages'] != '']

# 5. Se obtiene las 10 de los más populares
top_10_names= df_langs['cured_languages'].value_counts().head(10).index

#6. Se incluye Work_seniority, salary_monthly y language
df_final = df_langs[df_langs['cured_languages'].isin(top_10_names)].copy()

# 7. Visualización
visual = df_langs['cured_languages'].value_counts().head(10).reset_index()
visual.columns= ['lenguaje', 'frecuencia']
print("Los 10 lenguajes más populares son:")
print(visual)
```

c) **Definición de Rangos y Valores**


*   **Salary_monthly_neto:** seleccionamos el rango entre 300.000 y 10.000.000, porque los valores por debajo no representan el mercado, los valores por encima son outliers.

*  **work_seniority:** seleccionamos solo las tres principales categorías: Junior, Semi-Senior y Senior, para estandarizar la comparación.





```
top_10_names = df_langs['cured_languages'].value_counts().head(10).index
df_final = df_langs[df_langs['cured_languages'].isin(top_10_names)].copy()

# (Selección de filas por Salario y Seniority) ---
min_salario = 300000
max_salario = 10000000
experiencia_valida = ['Junior', 'Semi-Senior', 'Senior']

# Creamos df_analisis que es la tabla final
df_analisis = df_final[
    (df_final['salary_monthly_NETO'] >= min_salario) &
    (df_final['salary_monthly_NETO'] <= max_salario) &
    (df_final['work_seniority'].isin(experiencia_valida))
].copy()

print("Resumen de la sub-población seleccionada:")
print(df_analisis['work_seniority'].value_counts())
```

### Conteo de frecuencias de los lenguajes de programación

La columna que contiene información sobre los lenguajes de programación utilizados es `tools_programming_languages`. Sus valores son strings con los lenguajes seleccionados separados por comas.



```
df.tools_programming_languages[:3]
```

Las siguientes celdas de código separan estos lenguajes de programación y cuentan la frecuencia con la que aparecen.

No es necesario entender este código en profundidad, aunque sí es un buen ejercicio.

```
# Convert the comma-separated string of languages to a list of string.
# Remove 'ninguno de los anteriores' option, spaces and training commas.
def split_languages(languages_str):
  if not isinstance(languages_str, str):
    return []
  # Remove 'other' option
  languages_str = languages_str.lower()\
    .replace('ninguno de los anteriores', '')
  # Split string into list of items
  # Remove spaces and commas for each item
  return [lang.strip().replace(',', '')
          for lang in languages_str.split()]

# Create a new column with the list of languages
df.loc[:, 'cured_programming_languages'] = df.tools_programming_languages\
    .apply(split_languages)
if 'cured_programming_languages' not in relevant_columns:
    relevant_columns.append('cured_programming_languages')

# Duplicate each row of df for each programming language
# mentioned in the response.
# We only include in df_lang the columns we are going to analyze later, so we
# don't duplicate innecesary information.
df_lang = df.cured_programming_languages\
    .apply(pd.Series).stack()\
    .reset_index(level=-1, drop=True).to_frame()\
    .join(df[relevant_columns])\
    .rename(columns={0: 'programming_language'})
# Horrible programming style! But a lot of data science code can be written with
# as concatenations of functions (pipelines), and there's no elegant way of
# doing that on Python.
df_lang[:5]
```

En la columna `programming_language` se encuentra cada lenguaje por separado. Notar que si una respuesta contenía 3 lenguajes, como `"HTML, Javascript, Python"`, la fila ha sido replicada 3 veces. Por ello, hay tres filas con índice 1.

```
language_count = df_lang.programming_language.value_counts()\
    .reset_index()\
    .rename(columns={'index': 'language', 'programming_language': 'frequency'})
language_count[:10]
```
